```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
print os.path
if os.path.isfile("RCI_OffenceByMonth_1995-2012.csv"):
    filepath = "downloads/RCI_OffenceByMonth_1995-2012.csv"
    print "loading from file"
else:
    filepath = "http://data.gov.au/storage/f/2013-09-12T23%3A32%3A36.918Z/rci-offencebymonth.csv"
    print "loading from the internet"
RCI_OffenceByMonth_9512 = pd.read_csv(filepath)
print "done"
```

```python
# Getting rid of all whitespace in months-years
RCI_OffenceByMonth_9512.columns = [x.strip().replace(" ", "_") for x in RCI_OffenceByMonth_9512.columns]
test_df = RCI_OffenceByMonth_9512
```

```python
test_df.columns
```

```python
#Taking columns Jan_1995 to Dec_1995 as they make up the year of 1995
years95 = test_df.iloc[:, 4:16]
years95['LGA'] = test_df.iloc[:, 1]
years95
```

```python
#Summing up all crimes in each area of each month in 1995
muppet95 = years95.groupby('LGA').sum()
muppet95
```

```python
#Summing up the total of all months in each area
clown95 = muppet95.sum(axis=1)
clown95
```

```python
#Sorting each area in values from highest to lowest
newcolumnssorted95 = clown95.sort_values(ascending=False)
newcolumnssorted95
```

```python
#Collecting the Top Ten criminal areas in 1995
topten95 = newcolumnssorted95.iloc[:10]
topten95
```

```python
#Bar chart of the Top Ten Criminal Areas in 1995
topten95graph = topten95.plot.bar(x='LGA', y='Jan_1995', facecolor = 'cyan')
plt.title('Top 10 Criminal Areas 1995', fontsize=20)
plt.xlabel('Cities', fontsize=16)
plt.ylabel('Number of Crimes', fontsize=16)
topten95graph
```

```python
muppet12 = muppet12.reset_index()
muppet12
```

```python
muppet12["Total"] = muppet12.sum(axis = 1)
muppet12
```

```python
if os.path.isfile("population_nsw.xlsx"):
    filepath = "population_nsw.xlsx"
    print "loading from file"

population_data = pd.read_excel(filepath)
print "done"
```

```python
population_data["Local Government Area"], population_data["Drop"] = population_data["Local Government Area"].str.split("(", 1).str
population_data = population_data.drop("Drop", 1)
population_data["Local Government Area"] = population_data["Local Government Area"].str.strip()
population_data = population_data.rename(columns = {"Local Government Area": "LGA"})
population_data.head()
```

```python
population12 = population_data.iloc[:, [1, 8]]
population12.head()
```

```python
crime_ppls_12 = pd.concat([muppet12, population12], axis=1, join='inner')
crime_ppls_12
```

```python
crime_column = list(muppet12)
crime_ppls_12 = crime_ppls_12.rename(columns = {"2012r": "Population"})
crime_ppls_12[crime_column[1:]] = crime_ppls_12[crime_column[1:]].div(crime_ppls_12["Population"], axis=0)
crime_ppls_12
```

```python
#Bar chart of the Top Ten Criminal Areas in 2012
crime_ppls_12_sorted = crime_ppls_12.sort_values(by=['Total'], ascending=0)
top10crime2012 = crime_ppls_12_sorted.iloc[:10]

top10crime2012.plot.bar(x='LGA', y='Total', facecolor = 'magenta')
plt.title('Top 10 Criminal Area 2012 (per population)', fontsize=20)
plt.xlabel('Cities', fontsize=16)
plt.ylabel('Number of Crimes', fontsize=16)
```
