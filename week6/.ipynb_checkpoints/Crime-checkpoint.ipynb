# Fire vs. Crime incidents

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import geopandas as gp
import shapely
import fiona

%matplotlib inline
plt.rcParams['figure.figsize'] = (20, 10)
saved_style_state = matplotlib.rcParams.copy()
```

```python
if os.path.isfile("population_nsw.xlsx"):
    filepath = "population_nsw.xlsx"
    print "loading from file"

population_data = pd.read_excel(filepath)
print "done"
```

```python
shp_file_name = "NSW_LOCALITY_POLYGON_shp.shp"
zip_file_name = "nswlocalitypolygonshp.zip"
web_file_path = ("http://data.gov.au/dataset/91e70237-d9d1-4719-a82f-e71b811154c6/resource/" 
                 "bf8b5180-fcea-44bd-bd76-af14cc4b0fe0/download/nswlocalitypolygonshp.zip")
```

```python
def unzip_zipfile(zipped_file_path, put_it_here="."):
    import zipfile
    zip_of_suburbs = zipfile.ZipFile(zipped_file_path, 'r')
    zip_of_suburbs.extractall(put_it_here)
    zip_of_suburbs.close()
```

```python
# Get the data loaded. This is a bit complicated because it's online as a zip file.
# If we don't have it right now, we need to get it and unzip it.
if os.path.isfile(shp_file_name):
    print "loading from file"
else:
    if os.path.isfile(zip_file_name):
        print "unzipping"
        unzip_zipfile(zip_file_name)
    else:
        import requests
        print "loading from the internet"
        page = requests.get(web_file_path)
        with open(zip_file_name, 'wb') as z:
            z.write(page.content)
        unzip_zipfile(zip_file_name)

print "done"

```

```python
population_data["Local Government Area"], population_data["Drop"] = population_data["Local Government Area"].str.split("(", 1).str
population_data = population_data.drop("Drop", 1)
population_data["Local Government Area"] = population_data["Local Government Area"].str.strip()
```

```python
if os.path.isfile("Reported crime offence by month.xlsx"):
    filepath = "Reported crime offence by month.xlsx"
    print "loading from file"

crime_data = pd.read_excel(filepath)
print "done"
```

```python
crime_data.drop(crime_data.columns[range(3, 255)], axis=1, inplace=True)
crime_data["Total in 2016"] = crime_data.sum(axis=1)
```

```python
if os.path.isfile("type-of-incident-by-local-government-area.xlsx"):
    filepath = "type-of-incident-by-local-government-area.xlsx"
    print "loading from file"

fire_data = pd.read_excel(filepath)
print "done"
```

```python
fire_data["Total incidents except calls"] = fire_data["Fire, explosions"] + fire_data["Non-fire rescue calls including animal rescues"] + fire_data["Hazardous material incidents and other hazardous conditions"] + fire_data["Storm, floods and other natural disasters and calls for assistance from other agencies"]
```

```python
total_incident = fire_data.iloc[:, [0, 11]]
total_incident_sorted = total_incident.sort_values(by=['Total incidents except calls'], ascending=0)
total_incident_sorted.plot(x = 'Local Government Area', y = 'Total incidents except calls', kind = 'bar')
```

```python
fire_data = fire_data.sort_values(by=["Local Government Area"], ascending=1)
crime_data = crime_data.sort_values(by=["LGA"], ascending=1)
crime_data = crime_data.rename(columns = {"LGA": "Local Government Area"})
```

```python
#result = pd.concat([fire_data, crime_data], axis=1, join='inner')
```

```python
crime_lga = crime_data.groupby(crime_data["Local Government Area"]).sum()
crime_lga = crime_lga.reset_index()

total_crime = crime_lga.iloc[:, [0, -1]]
```

```python
total_population = population_data.iloc[:, [1, -1]]
```

```python
total_data = pd.concat([total_population, total_crime, fire_data], axis=1, join='inner')
total_data
```

```python
total_data = total_data.rename(columns = {"Total in 2016": "Total Crimes", "2016p": "Population"})
total_data
```

```python
fire_column = list(fire_data)
total_data[fire_column[1:]] = total_data[fire_column[1:]].div(total_data["Population"], axis=0)
total_data["Total Crimes"] = total_data["Total Crimes"].div(total_data["Population"], axis=0)
total_data
```

```python
y_crime_fire_sorted = total_data.iloc[:, [0, 3, -1]]
y_crime_fire_sorted = y_crime_fire_sorted.sort_values(by=['Total Crimes'], ascending=0)
y_crime_fire_sorted.plot(x="Total Crimes", y="Total incidents except calls", kind = "scatter")

y_crime_fire_sorted.head(10)
```

```python
y_crime_fire_sorted.plot(x="Local Government Area", y="Total Crimes", kind = "bar")
```

```python
y_crime_fire_sorted = y_crime_fire_sorted.sort_values(by=['Total incidents except calls'], ascending=0)
y_crime_fire_sorted.plot(x="Local Government Area", y="Total incidents except calls", kind = "bar")
```

```python
y_crime_fire_sorted["Fire-Crime"] = y_crime_fire_sorted["Total incidents except calls"].div(total_data["Total Crimes"], axis=0)
y_crime_fire_sorted = y_crime_fire_sorted.sort_values(by=['Fire-Crime'], ascending=0)
y_crime_fire_sorted.head(10)
y_crime_fire_sorted.plot(x="Local Government Area", y="Fire-Crime", kind = "bar")
y_crime_fire_sorted
```

```python
y_fire_crime_sorted = y_crime_fire_sorted.sort_values(by=['Total Crimes'], ascending=0)
y_fire_crime_sorted.head(10)
ax = y_fire_crime_sorted.plot(x="Total Crimes", y="Total incidents except calls", kind = "scatter")

y_fire_crime_sorted
```

```python
y_crime_fire_sorted['Local Government Area'] = y_crime_fire_sorted['Local Government Area'].str.upper()
y_crime_fire_sorted
```

```python
burbs = gp.GeoDataFrame.from_file(shp_file_name)
burbs.drop(["NSW_LOCA_1", "NSW_LOCA_3", "NSW_LOCA_4", "DT_RETIRE"], axis=1, inplace=True)
burbs
```

```python
burbs = burbs.rename(columns = {"NSW_LOCA_2": "Local Government Area"})
```

```python
burbs_data = pd.concat([burbs, y_crime_fire_sorted], axis=1, join='inner')
burbs_data
```

```python
burbs_data = burbs_data.rename(columns = {"Local Government Area": "Local_Government_Area"})
```

```python
right_here = shapely.geometry.point.Point(151.2299732, -33.9178754)
burbs_data["distance_from_UNSW"] = burbs_data.geometry.distance(right_here)
burbs["distance_from_UNSW"] = burbs.geometry.distance(right_here)

base=burbs[burbs.distance_from_UNSW<0.2].plot(color='white')
base2=burbs_data[burbs_data.distance_from_UNSW<0.2].plot(ax=base, color='white')
wanted_burbs = burbs_data[burbs_data.distance_from_UNSW<0.2]
base3=wanted_burbs[wanted_burbs["Total Crimes"]>0.1].plot(ax=base2, color='red')
wanted_burbs[wanted_burbs["Total incidents except calls"]>0.1].plot(ax=base3, color='blue')

for idx, row in wanted_burbs.iterrows():
    plt.annotate(s=row.Local_Government_Area, 
                 xy=(2, 1),
                 horizontalalignment='center')
plt.title("High Fire/Crime Incidents Area");
```
